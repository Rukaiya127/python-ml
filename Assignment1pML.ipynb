```python
#2
def safe_divide(a , b):
 try:
  return a/b
 except ZeroDivisionError:
  return "Division by zero"
a , b = map(int, input().split())
ans = safe_divide(a , b)
print(ans)
```

```python
#3
import math

class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return math.pi * self.r * self.r


n = input().split()
shape_type = n[0]

if shape_type == "Rectangle":
    w = float(n[1])
    h = float(n[2])
    shape = Rectangle(w, h)
elif shape_type == "Circle":
    r = float(n[1])
    shape = Circle(r)

print(f"Area: {shape.area():.2f}")
```

```python
#4
sent = input().split()
longst = sent[0]

for word in sent:
    if len(word) > len(longst):
        longst = word

print(longst)
```

```python
#5
import math

n = int(input())
temp = n
sum_fact = 0

while temp > 0:
    digit = temp % 10
    sum_fact += math.factorial(digit)
    temp //= 10

if sum_fact == n:
    print("Strong Number")
else:
    print("Not Strong Number")
```
